## Вопросы

### 1. Что такое нулевая гипотеза? Если вы пытаетесь найти новую частицу, то что будет нулевой гипотезой в вашем эксперименте?


Нулевая гипотеза - это предположение, истинность которого по отношению к другим гипотезам проверяется в эксперименте. Подтверждение или опровержение гипотезы основывается на результатах экспреримента. При помощи некоторого критерия вычисляется, на сколько вероятно получить такие данные при условии, что эта гипотеза верна.

Нулевой гипотезой в эксперименте с поиском новой частицы будет предположение, что новая частица отсутствует. То есть, если после проведения эксперимента была принята нулевая гипотеза, то это значит, что результаты эксперимента с определенной вероятностью согласуются с моделью, где новой частицы нет. В случае отклонения нулевой гипотезы мы предполагаем, что новая частица существует.

### 2. Что показывает уровень значимости для критической области статистики?


Уровень значимости показывает, с какой вероятностью будет принята альтернативная гипотеза при условии, что верна нулевая гипотеза. Обычно это столь малая величина, что событие с такой вероятностью является практически невозможным.

Если статистика критерия принимает значение, равное или более экстремальное, чем уровень значимости, то нулевая гипотеза отклоняется.

### 1. Muriel Bristol утверждала, что может сказать, чай или молоко было добавлено в чашку сначала. Ronald Fisher решил провести эксперимент: предоставляется 8 чашек, в 4 наливают сначала молоко, в других 4 - чай; нужно определить 4 чашки одного типа. При каком результате можно будет сказать, что такая способность действительно есть? С чем могут возникнуть проблемы?

Нулевая гипотеза - что Muriel Bristol не может отличить, чай или молоко было добавлено первым. То есть она выбирает ответ рандомно, и вероятность того, что ее ответ совпадет с истинным, будет равна p=0.5. Альтернативная гипотеза состоит в том, что выбор не случаен и p>0.5.

Рассмотрим вероятности выбрать правильно 4, 3, 2, 1 или 0 чашек правильно при верной нулевой гипотезе.

Всего существует $C_{8}^4 = \frac{8!}{4!4!} = 70$ способов выбрать 4 чашки. Пусть $P(k)$ - вероятность, что среди этих 4 чашек $k$ выбрано корректно:
1) $P(0) = \frac{1}{70} = 0.01428$

2) $P(1) = \frac{C_{1}^4 * C_{3}^4}{70}  = \frac{\frac{4!}{3!} * \frac{4!}{3!}}{70} = \frac{16}{70} = 0.2285$

3) $P(2) = \frac{C_{2}^4 * C_{2}^4}{70}  = \frac{\frac{4!}{2! 2!}^2}{70} = \frac{36}{70} = 0.51$

4) $P(3) = \frac{C_{3}^4 * C_{1}^4}{70}  = \frac{\frac{4!}{3!}^2}{70} = \frac{16}{70} =  0.2285$

5) $P(4) = \frac{1}{70} = 0.01428$


При проведении эксперимента в качестве pvalue можно взять эту статистику P(k), где k - количество угаданных чашек в эксперименте. 

Если в качестве критического значения взять $\alpha=0.05$, то нулевая гипотеза будет отвергнута только в одном случае - если k=4, то есть все чашки угаданы верно.



Проблема может заключаться в том, что разница вероятностей выбрать правильно 3 и 4 чашки огромная, поэтому нулевая гипотеза будет принята, даже если заявленные способности есть, но была допущена одна ошибка.
Более достоверным будет эксперимент с большим количеством испытаний.

### 2. В 2020 году средний балл ЕГЭ по физике составил 54.5 балла, экзамен сдали 139 574 школьника. В 2019 году средний балл - 54,4, а сдавало 142 607 школьника. Является ли это увеличение среднего балла статистически значимым? (Прим. распределение баллов нормальное со среднеквадратичным разбросом - 6,1 балл; для подсчета статистической значимости можно использовать критерий Стьюдента)

Используем проверку гипотезы о равенстве математических ожиданий двух распределений: 

$F_{2020} = N(m_{0}, \sigma )$ и $F_{2019} = N(m_{1}, \sigma )$ , где $\sigma = 6.1 $

Тогда гипотезы выглядят следующим образом:

$H0: m_0 = m_1$

$H1: m_0 \neq m_1$

Пусть необходимый уровень значимости будет равен $\alpha = 0.05$

Используем критерий Стьюдента (мы можем его применить, так как речь идет о нормально распределенных случайных величинах): $Z = \frac{\bar X - \bar Y}{ \sqrt{\frac{\sigma^2}{m} + \frac{\sigma^2}{n}} }$ будет починяться нормальному распределению N(0, 1) если H0 верна. 

$\bar X = 54.5$

$\bar Y = 54.4$

$m = 139574$

$n = 142607$

$Z = \frac{54.5 - 54.4}{ \sqrt{\frac{6.1^2}{139574} + \frac{6.1^2}{142607}} } = 6,12$

$pvalue = 2*min(F_t(Z), 1 - F_t(Z))$

```python
import numpy as np
from scipy.stats import norm

X = 54.4
Y = 54.5
m = 139574
n = 142607
D = 6.1**2

Z = (X - Y)/np.sqrt(D/m + D/n)
pvalue = 2 * min(norm.cdf(Z), 1 - norm.cdf(Z))
print(pvalue)
```

Pvalue существенно меньше уровня значимости $\alpha$, поэтому мы отвергаем нулевую гипотезу в пользу альтернативной. То есть, увеличение среднего балла является статистически значимым.

### 4. В классической пачке m&m's 40 конфет, среди них в среднем красных - 13%, желтых - 14%, оранжевых - 21%, зеленых - 20%, коричневых - 12% и синих - 21%. Проверьте гипотезу, что распределение цветов конфет равномерное.

Равномерное распределение цветов эквивалентно подтверджению того, что цвета подчиняются мультиномиальному распределению с 6 исходами с равными вероятностями: $p_i=\frac{1}{6}$ для всех цветов.

H0: $p_i=\frac{1}{6}$

H1: $p_i \neq \frac{1}{6}$

Для проверки гипотезы можно использовать критерий Пирсона (или $\chi^2$). Его статистика равна:
$$
Z = \sum \frac{(n_i - N*p_i)^2}{N*p_i}
$$

$n_i$ - среднее число наблюдений i-го цвета

$p_i$ - теоретическая вероятность наблюдать i-ый цвет

$N$ - количество конфет

В качестве уровня значимости возьмем $\alpha = 0.05$.

Статистика критерия приблизительно равна распределению $\chi^2$ с 5-ю степенями свободы (количество цветов - 1). Используем его для проверки гипотезы, $pvalue = 1 - F_{\chi^2}(Z)$

```python
import scipy

N = 40
expected_prob = 1 / 6
probs = [0.13, 0.14, 0.21, 0.2, 0.12, 0.21]
expected = N * expected_prob

Z = sum([np.power(N*prob - expected, 2)/ expected for prob in probs])
pvalue = 1 - scipy.stats.chi2.cdf(x=Z, df=5)


print(f"Statistic: {Z}, pvalue: {pvalue}")
```

Полученное pvalue существенно больше, чем $\alpha=0.05$, поэтому мы не можем отвергнуть нулевую гипотезу. Значит, распределение конфет равномерное.

### 5. Монетку подбросили 1000 раз, 550 раз выпал орел. Как вы думаете, монетка честная? Почему?


Проверим гипотезу о параметре биномиального распределения, где p - вероятность того, что выпал орел. При проведении N=1000 экспериментов мат ожидание количества выпавших орлов (при условии, что монета честная) равно $\mu = 1000*0.5 = 500$, а получено было $k = 550$. Проверим гипотезу, что орел выпадает с большей вероятностью.

HO: p=0.5, то есть монета честная

H1: p > 0.5, то есть монета нечестная, орел выпадает чаще

В качестве уровня значимости возьмем $\alpha = 0.05$
 
В качестве проверочной статистики возьмем вероятность того, что случайная вечинина примет значение, большее или меньшее k=550, т.е. $pvalue=P(X>=550)$



```python
import scipy

p = 0.5
x = 550
N = 1000

p_value = 1 - scipy.stats.binom.cdf(x, N, p)
print(f'p-value: {p_value} for alpha=0.05')
```

Так как p-value принимает значение, меньшее уровня значимости, то мы можем отвергнуть нулевую гипотезу. Значит, монетка не честная.
